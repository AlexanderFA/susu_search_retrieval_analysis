```python
members = df.groupby(['Тип олимпиады', 'Класс']).agg('count')
members
```

```python
df_moscow = df[df['Тип олимпиады'] == 'Московская олимпиада']
moscow_class = df_moscow.groupby('Класс').agg('count')
moscow_class
```

```python
moscow_class['global_id'].plot(kind='bar', xlabel='Московская ' + moscow_class.index.astype(str))
```
